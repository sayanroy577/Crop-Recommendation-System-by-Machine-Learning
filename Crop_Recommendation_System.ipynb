# Importing Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Upload Dataset

```python
df=pd.read_csv("Crop_recommendation.csv")
```

# Drop Unnecessary Column

```python
df.drop(columns=['Unnamed: 8','Unnamed: 9'],inplace=True)
```

# Splitting the data into X & Y

```python
x=df.drop(columns='label')
y=df['label']
```

**Dividing the data into training & testing**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

**Logistic Regression Model Training**

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

**Prediction on Testing Data**

```python
x_test_pred=lr.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy Score is :",accu_score)
```

# Decision Tree model training

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

**Prediction of testing data**

```python
x_test_pred=dt.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy Score is :",accu_score)
```

# Random Forest Classifier Model Training

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

**Prediction of testing data**

```python
x_test_pred=rf.predict(x_test)
accu_score=accuracy_score(x_test_pred,y_test)
print("Accuracy Score is :",accu_score)
```
